```python
import torch
import torch.nn as nn

# Define LeNet input layer
class LeNetInput(nn.Module):
    def __init__(self):
        super(LeNetInput, self).__init__()
        self.input_layer = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1, padding=0)  # First Conv Layer

    def forward(self, x):
        return self.input_layer(x)

# Create model
model = LeNetInput()
x = torch.randn(1, 1, 32, 32)  # Sample input (batch_size=1, channels=1, height=32, width=32)
output = model(x)
print("Output shape:", output.shape)
```
